# Part 1

```python
# TODO define gaussian function here
# (I defined two)
def stdNorm(windowSize, ptSpacing, stdDev, height):
    numPts = windowSize/ptSpacing
    normStdDev = stdev/(ptSpacing**-1.)
    stdNormal = scp.signal.gaussian(numPts, normStdDev)
    return(stdNormal)

def gaussian(x, mean, stdev, scaling):
    return (float(stdev)**-1.)*((2*np.pi)**-.5)*(np.exp(-.5*(((x - mean)/float(stdev))**2)))*scaling
```

```python
# TODO plot gaussians here
x = np.arange(-5, 5, 0.01)
plt.plot(x, gaussian(x, 0, 1, 1), 'black', label = 'Standard Normal')
plt.plot(x, gaussian(x, 0, .6, 1), 'blue', label = 'stddev = 0.6')
plt.plot(x, gaussian(x, 0, .8, 1), 'darkblue', label = 'stddev = 0.8')
plt.plot(x, gaussian(x, 0, 2, 1), 'red', label = 'stddev = 2')
plt.plot(x, gaussian(x, 0, 1.5, 1), 'darkred', label = 'stddev = 1.5')
plt.plot(x, gaussian(x, 2.5, 1, 1), 'green', label = 'mean = 2.5')
plt.plot(x, gaussian(x, 2.5, 1, 1.2), 'darkgreen', linestyle = 'dashed', label = 'mean = 2.5, scaling = 1.1')
plt.plot(x, gaussian(x, -2.5, 1, 1), 'orange', label = 'mean = -2.5')
plt.plot(x, gaussian(x, -2.5, 1, 0.8), 'orange', linestyle = 'dashed', label = 'mean = -2.5, scaling = 0.9')
plt.legend(frameon = True, bbox_to_anchor=(1.05, 1), loc=2)
plt.show()
```

```python
# TODO fit gaussians, with known parameters, here
data_Y = gaussian(x, -2.5, 1, 1)
data_X = x
fitParams, fitCov = scp.optimize.curve_fit(gaussian, data_X, data_Y)

# TODO plot result
plt.plot(data_X, data_Y, 'dimgrey', lw=5, label='Data')
plt.plot(data_X, gaussian(x, -2.5, 1, 1), 'orange', linestyle='dashed', label='Fitted Curve')
plt.legend(frameon = True, bbox_to_anchor=(1.05, 1), loc=2)
plt.text(1, 0.25, 'The fitted Mean is ' + str(fitParams[0]) + '\nThe fitted stdev is ' + str(fitParams[1]))
plt.show()
```

```python
# TODO: load data here
HDA_data = np.loadtxt('HDAlpha_data')
```

```python
# TODO: plot it
Freq = HDA_data[:, 0]
Amp = HDA_data[:, 1]
plt.plot(Freq, Amp)
plt.show()
```

```python
# TODO: plot data as emission signal
plt.plot(Freq, -Amp)
plt.show()
```

```python
# TODO: fit a function to this data and plot it
def doubleGaussian(x, mean1, stdev1, scale1, mean2, stdev2, scale2):
    return gaussian(x, mean1, stdev1, scale1) + gaussian(x, mean2, stdev2, scale2)

fitParams, fitCov = scp.optimize.curve_fit(doubleGaussian, Freq, -Amp, [13127, 1, 0.001, 13131, 1, 0.001])

plt.plot(Freq, -Amp, 'dimgrey', lw=4, label = "Data")
plt.plot(Freq, gaussian(Freq, fitParams[0], fitParams[1], fitParams[2]), 'red', lw=1, label = "First fitted Gaussian distribution")
plt.plot(Freq, gaussian(Freq, fitParams[3], fitParams[4], fitParams[5]), 'blue', lw=1, label = "Second fitted Gaussian distribution")
plt.legend(frameon = True, bbox_to_anchor=(1.05, 1), loc=2)
plt.ticklabel_format(style='sci', axis='y', scilimits=(0,0))
plt.suptitle('Hydrogen/Deuterium Absorption Signal', fontsize = 14)
plt.title('With Gaussian fits', fontsize = 11)
plt.xlabel('Frequency')
plt.ylabel('Amplitude')
plt.text(13124.5, 1.7*(10**-5), 
         'mean =' + '{:.2E}'.format(Decimal(fitParams[0])) +
         '\nstdev =' + '{:.2E}'.format(Decimal(fitParams[1])) + 
         '\nscale =' + '{:.2E}'.format(Decimal(fitParams[2])), 
         fontsize = 7, color = 'red')
plt.text(13132.5, 1.6*(10**-5),
         'mean =' + '{:.2E}'.format(Decimal(fitParams[3])) +
         '\nstdev =' + '{:.2E}'.format(Decimal(fitParams[4])) + 
         '\nscale =' + '{:.2E}'.format(Decimal(fitParams[5])),
         fontsize = 7, color = 'blue')
plt.show()
```

```python
# Add double Gaussian fit
plt.plot(Freq, -Amp, 'dimgrey', lw=4, label = "Data")
plt.plot(Freq, doubleGaussian(Freq, fitParams[0], fitParams[1], fitParams[2], fitParams[3], fitParams[4], fitParams[5]), 
         'orange', lw=1, label = "Double Gaussian distribution")
plt.legend(frameon = True, bbox_to_anchor=(1.05, 1), loc=2)
plt.ticklabel_format(style='sci', axis='y', scilimits=(0,0))
plt.suptitle('Hydrogen/Deuterium Absorption Signal', fontsize = 14)
plt.title('With Double Gaussian fits', fontsize = 11)
plt.xlabel('Frequency')
plt.ylabel('Amplitude')
plt.text(13124.5, 1.7*(10**-5), 
         'First Gaussian Peak' +
         '\nmean =' + '{:.2E}'.format(Decimal(fitParams[0])) +
         '\nstdev =' + '{:.2E}'.format(Decimal(fitParams[1])) + 
         '\nscale =' + '{:.2E}'.format(Decimal(fitParams[2])), 
         fontsize = 7)
plt.text(13132.5, 1.6*(10**-5),
         'Second Gaussian Peak' +
         '\nmean =' + '{:.2E}'.format(Decimal(fitParams[3])) +
         '\nstdev =' + '{:.2E}'.format(Decimal(fitParams[4])) + 
         '\nscale =' + '{:.2E}'.format(Decimal(fitParams[5])),
         fontsize = 7)
plt.show()
```

```python
def flux_for_dipole(x_pos, y_pos, Q_vals, X, Y):
    """
    Returns the flux based on the information of the charges
    and the grid of the plot.
    Code modified from Physics 113, written by Blas Cabrera
    """
    flux = np.zeros((len(X), len(Y)))
    x_m = np.outer(np.ones((X.size)), X); 
    y_m = np.outer(Y, np.ones((Y.size))); 
    for n in range(len(Q_vals)):
        flux += 2*np.pi*Q_vals[n]*(1 - (x_pos[n]-x_m)/np.sqrt((y_pos[n]-y_m)**2+(x_pos[n]-x_m)**2))
    return flux
```

```python
# TODO create arrays of information
X_pos = [-5., 9.]
Y_pos = [-5., 5.]
Charge = [-4., 1.]

X_space = np.arange(-31.5, 31.5, 0.5)
Y_space = np.arange(-31.5, 31.5, 0.5)

# TODO get flux
flux = flux_for_dipole(X_pos, Y_pos, Charge, X_space, Y_space)

# TODO plot contours of the flux
plt.contour(X_space, Y_space, flux, 25)
plt.scatter(X_pos, Y_pos, color = 'black', s = [a*b*10 for a, b in zip(Charge, Charge)], zorder = 2)
plt.suptitle('Electric Field Contour Lines', fontsize = 13)
plt.title('for charges of -4 and 1', fontsize = 8)
plt.show()
```
